# Test the 1-D transform
For f(x) = {1, 4, -3, 0}, the transform is supposed to be {1, 4, -2.121, -2.121} for a 2-scale transform

```python
f = np.array([1,4,-3,0],dtype='float64')   #The type has to be float otherwise, 
transform = one_DDWT(f,2)
print(transform)
```

```python
reconstruct = inverse_one_DDWT(transform,2)
print(reconstruct)
```

We have the good transform and the good inverse transform

# Test the 2-D Transform

```python
image = cv2.imread('data/image.png', cv2.IMREAD_GRAYSCALE)
image = image.astype('float64')
print(np.shape(image))
plt.imshow(image,cmap='gray')
```

```python
transform = twoDDWT(image,2)
plot_transform(transform,2)
plt.savefig("data/2D_Transform.png")
```

```python
reconstruct = inverse_twoDDWT(transform,2)
print(np.sum(reconstruct-image)/(512*512))
plot_transform(reconstruct,0)
```

The result of the transform is what was expected. When computing the inverse we get the same image as the input. The only error are the approximations of the computation (e-16 is negligable)
